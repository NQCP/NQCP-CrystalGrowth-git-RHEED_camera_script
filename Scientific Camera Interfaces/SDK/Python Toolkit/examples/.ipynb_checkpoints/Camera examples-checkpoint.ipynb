```python
import cv2
import numpy as np

# Replace 'your_video.tiff' with the actual file path of your TIFF video.
video_path = 'image.tif'

# Create a VideoCapture object to read the video.
cap = cv2.VideoCapture(video_path)

# Initialize an empty list to store the frames.
frames = []

# Loop through the video and extract each frame.
while True:
    ret, frame = cap.read()
    
    # Break the loop when we reach the end of the video.
    if not ret:
        break
    
    # Append the frame to the frames list.
    frames.append(frame)

# Release the VideoCapture object.
cap.release()

# Convert the list of frames into a NumPy array.
video_array = np.array(frames)
video_array.shape
```

```python
plt.imshow(video_array[0]/4)
plt.show()
plt.hist(video_array[0].flatten(), bins= 50)
plt.show() 
plt.imshow(video_array[0]/np.max(video_array[0]))
```

```python
import imageio
import numpy as np

# Replace 'your_video.tiff' with the actual file path of your TIFF video.
video_path = 'image.tif'

# Read the TIFF video using imageio.
video = imageio.get_reader(video_path)

# Initialize an empty list to store frames.
frames = []

# Iterate through the frames in the TIFF video.
for frame in video:
    # Append each frame (which is already a NumPy array) to the frames list.
    frames.append(frame)

# Convert the list of frames into a NumPy array.
video_array = np.array(frames)

# Close the video reader.
video.close()
```

```python
frame_to_display = video_array[0]


alpha = 0.25  # Controls contrast (1.0 is no change)
beta = 1 # Controls brightness (0 is no change)

adjusted_frame = cv2.convertScaleAbs(frame_to_display, alpha=alpha, beta=beta)

# Display the adjusted frame using imshow.
plt.imshow(cv2.cvtColor(adjusted_frame, cv2.COLOR_BGR2RGB))
plt.title("Adjusted Frame 0")
plt.axis('off')  # Turn off axis labels and ticks
plt.show()
```

```python
import cv2
import numpy as np
from instrumental.drivers.cameras import uc480

# init camera
instruments = uc480.list_instruments()
cam = uc480.UC480_Camera(instruments[0])

# params
cam.start_live_video(framerate = "10Hz")

while cam.is_open:
     
    frame = cam.grab_image(timeout='100s', copy=True, exposure_time='10ms')
     
    frame1 = np.stack((frame,) * 3,-1) #make frame as 1 channel image
    frame1 = frame1.astype(np.uint8)

    gray = cv2.cvtColor(frame1, cv2.COLOR_BGR2GRAY)

     #now u can apply opencv features

    cv2.imshow('Camera', gray)
     
    if cv2.waitKey(30) & 0xFF == ord('q'):
        break

cam.close()
cv2.destroyAllWindows()
```
